```python
import torch
def cut(inp, zsize, xsize, ysize):
        #inp = unfold(inp)
        inp = inp.unfold(3, zsize , zsize ,)\
                 .unfold(1, xsize , xsize ,)\
                 .unfold(2, ysize , ysize ,)
        #print(inp.shape)
        #num =  inp.shape[1] / zsize * inp.shape[2] /xsize  * inp.shape[3] / ysize
        num =  inp.shape[1] * inp.shape[2] * inp.shape[3]
        #print(num)
        return torch.reshape(inp, (int(num), 1, zsize, xsize, ysize))
```
